**LIGHTNESS METHOD**


```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
img = mpimg.imread('/content/dog.jpeg')

def lightness(img):
  max_rgb = np.max(img[..., :3], axis=2)
  min_rgb = np.min(img[..., :3], axis=2)
  return (max_rgb + min_rgb) / 2

gray_img_01 = lightness(img)
print(gray_img_01[0, 0])
plt.imshow(gray_img_01, cmap='gray')
```

**AVERAGE COLOR RGB METHOD**

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
img = mpimg.imread('/content/dog.jpeg')

def avg_color(img):
  return np.mean(img[..., :3], axis=2)

gray_img_02 = avg_color(img)
print(gray_img_02[0, 0])
plt.imshow(gray_img_02, cmap='gray')
```

**LUMINOSITY METHOD**

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
img = mpimg.imread('/content/dog.jpeg')

def luminosity(img):
  return 0.2126 * img[..., 0] + 0.72 * img[..., 1] + 0.07 * img[..., 2]

gray_img_03 = luminosity(img)
print(gray_img_03[0, 0])
plt.imshow(gray_img_03, cmap='gray')
```
